```python
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from textblob import TextBlob
import nltk
nltk.download('stopwords')
```

```python
def preprocess_text(text):
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Convert to lowercase
    text = text.lower()
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text
```

```python
# Apply text preprocessing to the text columns
cool_df['history_text_cleaned'] = cool_df['history_text'].apply(preprocess_text)
cool_df['powers_text_cleaned'] = cool_df['powers_text'].apply(preprocess_text)
```

```python
relevant_words = ['cool', 'awesome', 'amazing', 'badass']

# TF-IDF Vectorization
vectorizer = TfidfVectorizer(vocabulary=relevant_words)
history_tfidf = vectorizer.fit_transform(cool_df['history_text_cleaned'])
powers_tfidf = vectorizer.fit_transform(cool_df['powers_text_cleaned'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Define the relevant words
relevant_words = ['cool', 'awesome', 'amazing', 'badass']

# TF-IDF Vectorization with the predefined vocabulary
vectorizer = TfidfVectorizer(vocabulary=relevant_words)

# Transform the history text and powers text using the vectorizer
history_tfidf = vectorizer.fit_transform(cool_df['history_text_cleaned'])
powers_tfidf = vectorizer.fit_transform(cool_df['powers_text_cleaned'])

# Calculate similarity scores between relevant words and history text
history_similarity_scores = cosine_similarity(history_tfidf, vectorizer.transform([' '.join(relevant_words)]))

# Calculate similarity scores between relevant words and powers text
powers_similarity_scores = cosine_similarity(powers_tfidf, vectorizer.transform([' '.join(relevant_words)]))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate "coolness" score for each superhero
coolness_scores = history_tfidf.sum(axis=1) + powers_tfidf.sum(axis=1)

# Normalize scores (optional)
normalized_scores = (coolness_scores - coolness_scores.min()) / (coolness_scores.max() - coolness_scores.min())

normalized_scores_array = np.array(normalized_scores)

# Convert the matrix to an array and flatten it
flattened_scores = np.array(normalized_scores).flatten()
```

```python
# Sort the superheroes based on "coolness" scores
sorted_indices = np.argsort(flattened_scores)[::-1]  # Sort indices in descending order
top_10_indices = sorted_indices[:10]  # Select the top 10 indices

# Get the names and scores of the top 10 coolest superheroes
top_10_names = cool_df['name'].iloc[top_10_indices]
top_10_scores = flattened_scores[top_10_indices]

# Plot the top 10 coolest superheroes
plt.figure(figsize=(10, 6))
plt.bar(top_10_names, top_10_scores, color = 'skyblue')
plt.xlabel('Superhero')
plt.ylabel('Coolness Score')
plt.title('Top 10 Coolest Superheroes')
plt.xticks(rotation=45, ha='right')  # Rotate labels for better readability
plt.tight_layout()
plt.show()
```
